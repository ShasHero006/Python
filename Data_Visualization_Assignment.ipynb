Q.5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

Ans.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a synthetic dataset with correlated features
np.random.seed(42)  # For reproducibility
n_samples = 100

# Create synthetic correlated data
x = np.random.normal(0, 1, n_samples)
y = x + np.random.normal(0, 0.1, n_samples)  # y is correlated with x
z = x - np.random.normal(0, 0.5, n_samples)  # z is inversely correlated with x

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Calculate the correlation matrix
corr = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Adding labels and title
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Q.3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.
 np.random.seed(20)
 data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
 }

 df = pd.DataFrame(data)


 Ans.

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Set the seed for reproducibility
np.random.seed(20)

# Generate the sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Pivot the data to create a matrix of sales across days and months
heatmap_data = pd.pivot_table(df, values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap using Plotly
fig = px.imshow(heatmap_data, title="Sales Variation Across Months and Days", labels={'color': 'Sales'})

# Show the plot
fig.show()
```

Q.4.  Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(x**2+y**2)

x = np.linspace(-5, 5, 100)

 y = np.linspace(-5, 5, 100)

 x, y = np.meshgrid(x, y)

 z = np.sin(np.sqrt(x**2 + y**2))

 data = {

 'X': x.flatten(),

 'Y': y.flatten(),

 'Z': z.flatten()

 }

 df = pd.DataFrame(data)

 Ans.

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Define the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid for x and y
x, y = np.meshgrid(x, y)

# Compute z values based on the function x + y
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])

# Add labels and title
fig.update_layout(title='3D Surface Plot of sin(x + y)',
                  scene=dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'),
                  )

# Display the plot
fig.show()
```

Q.2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

Ans.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(10, 50, size=n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Create a DataFrame
data = {
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
}
df = pd.DataFrame(data)

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(df)

# Create a scatter plot
output_notebook()  # Use this to display the plot in a Jupyter Notebook
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter points
p.circle('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

# Show the plot
show(p)
```

Q.4. Create a Bokeh histogram to visualize the distribution of the given data.

 data_hist = np.random.randn(1000)

 hist, edges = np.histogram(data_hist, bins=30)

 Ans.

```python
# Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Generate random data
np.random.seed(42)  # For reproducibility
data_hist = np.random.randn(1000)  # Generate 1000 random numbers from a standard normal distribution

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)  # Compute the histogram

# Create a Bokeh histogram plot
output_notebook()  # Use this to display the plot in a Jupyter Notebook
p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value',
           y_axis_label='Frequency',
           height=400)

# Add the bars to the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", alpha=0.7)

# Show the plot
show(p)
```

Q.5.  Create a Bokeh heatmap using the provided dataset.

 data_heatmap = np.random.rand(10, 10)

 x = np.linspace(0, 1, 10)

 y = np.linspace(0, 1, 10)

 xx, yy = np.meshgrid(x, y)



 Ans.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
import numpy as np

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Reshape data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'values': data_heatmap.flatten()
}
source = ColumnDataSource(data)

# Create color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create Bokeh plot
p = figure(title="Heatmap", x_axis_label="x", y_axis_label="y",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           toolbar_location=None, tools="")  # Remove toolbar and tools

# Add heatmap rectangles
p.rect(x="x", y="y", width=0.1, height=0.1, source=source,
       fill_color={'field': 'values', 'transform': color_mapper},
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
